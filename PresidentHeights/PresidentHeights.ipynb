```python
import pandas as pd
import numpy as np
read_file = pd.read_csv("president_heights.csv")
```

```python
print(read_file.head())
print(read_file)
```

```python
height = np.array(read_file["height(cm)"])
```

```python
print("Mean of height ",height.mean())
```

```python
print("Std of height ",height.std())
```

```python
print("Max height= ",height.max())
print("Min height= ",height.min())
```

```python
print(read_file.describe())
print("Median ",np.median(height))
```

```python
import matplotlib.pyplot  as plt
import seaborn as sns
sns.set()
plt.hist(height)
plt.title("Height distribution of Presidents of USA")
plt.xlabel("height")
plt.ylabel("Number of persons")
plt.show()
plt.plot(height)
```
